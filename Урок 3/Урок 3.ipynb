4. Задание (в программе)

Решите систему уравнений:
y = x2 – 1
exp(x) + x∙(1 – y) = 1


```python
from scipy.optimize import fsolve, root

x = np.linspace(-5, 5, 1000)
fx = np.exp(x)+x*(2-x**2) - 1

def fx_(x):
    return np.exp(x)+x*(2-x**2) - 1

def y_(x):
    return x**2 - 1

plt.figure(figsize=(8,8))
plt.xlim(-5,8)
plt.ylim(-4,22)

x0 = [-1.5, 0, 3, 4]
x_root = np.array([fsolve(fx_, i) for i in x0])
y_root = np.array([fx_(i) for i in x_root])

plt.plot(x, fx)


y1 = x**2 - 1
y2 = (np.exp(x) + x - 1)/x

plt.plot(x, y1)
plt.plot(x, y2)

plt.grid(True)
print('Корнями системы уравнений являются: {}'.format(np.concatenate(x_root)))
```

Решите систему уравнений и неравенств:
y = x2 – 1
exp(x) + x∙(1 – y) > 1

```python
from scipy.optimize import fsolve, root

x = np.linspace(-5, 5, 1000)
fx = np.exp(x)+x*(2-x**2) - 1

def fx_(x):
    return np.exp(x)+x*(2-x**2) - 1

def y_(x):
    return x**2 - 1

plt.figure(figsize=(8,8))
plt.xlim(-5,8)
plt.ylim(-4,22)

x0 = [-1.5, 0, 3, 4]
x_root = np.array([fsolve(fx_, i) for i in x0])
y_root = np.array([fx_(i) for i in x_root])

plt.plot(x, fx)


y1 = x**2 - 1
y2 = (np.exp(x) + x - 1)/x

plt.plot(x, y1)
plt.plot(x, y2)

plt.grid(True)
print('Решением системы уравнений и неравенств является интервалы значений x: {} - {}, {} - {}, {} - {}'.format(float("-inf"), x_root[0], x_root[1], x_root[2], x_root[3], float("inf")))
```
